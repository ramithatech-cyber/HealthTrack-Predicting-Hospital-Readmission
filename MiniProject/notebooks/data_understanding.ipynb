This notebook covers initial exploration of the patient dataset:
- Loading and inspecting the data
- Summary statistics
- Class balance check
- Distribution plots for key features

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid', font_scale=0.95)
pd.set_option('display.max_columns', None)
print('Libraries loaded.')
```

## 1. Load Dataset

```python
df = pd.read_csv('../dataset/raw_data/patient_data.csv')
print(f'Shape: {df.shape}')
df.head()
```

## 2. Basic Info & Data Types

```python
df.info()
```

```python
# Check for missing values
print('Missing values:')
print(df.isnull().sum())
```

## 3. Summary Statistics

```python
num_cols = ['age', 'num_prior_visits', 'length_of_stay', 'num_medications', 'num_comorbidities']
df[num_cols].describe().round(2)
```

## 4. Target Variable – Class Balance

```python
counts = df['readmitted_30days'].value_counts()
print(counts)
print(f'\nReadmission rate: {counts[1]/len(df)*100:.1f}%')

fig, ax = plt.subplots(figsize=(5, 3.5))
ax.bar(['Not Readmitted', 'Readmitted'], counts.values, color=['#4a7fa5', '#b94a48'], width=0.45, edgecolor='white')
ax.set_title('Readmission Distribution', fontsize=11)
ax.set_ylabel('Count')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
for i, v in enumerate(counts.values):
    ax.text(i, v + 15, str(v), ha='center', fontsize=9)
plt.tight_layout()
plt.savefig('../outputs/graphs/readmission_distribution.png', dpi=150)
plt.show()
```

## 5. Age Distribution by Readmission

```python
fig, ax = plt.subplots(figsize=(6, 3.5))
for label, val, color in [('Not Readmitted', 0, '#4a7fa5'), ('Readmitted', 1, '#b94a48')]:
    ax.hist(df[df['readmitted_30days'] == val]['age'], bins=20, alpha=0.6, label=label, color=color, edgecolor='white')
ax.set_xlabel('Age')
ax.set_ylabel('Count')
ax.set_title('Age Distribution by Readmission Status', fontsize=11)
ax.legend()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()
plt.savefig('../outputs/graphs/age_distribution.png', dpi=150)
plt.show()
```

## 6. Categorical Feature Analysis

```python
# Readmission rate by diagnosis
diag_rate = df.groupby('diagnosis')['readmitted_30days'].mean().sort_values(ascending=False)
print('Readmission rate by diagnosis:')
print(diag_rate.round(3))

fig, ax = plt.subplots(figsize=(7, 4))
ax.bar(range(len(diag_rate)), diag_rate.values * 100, color='#b94a48', alpha=0.75, edgecolor='white')
ax.set_xticks(range(len(diag_rate)))
ax.set_xticklabels(diag_rate.index, rotation=30, ha='right', fontsize=9)
ax.set_ylabel('Readmission Rate (%)')
ax.set_title('Readmission Rate by Diagnosis', fontsize=11)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()
plt.savefig('../outputs/graphs/diagnosis_readmission.png', dpi=150)
plt.show()
```

## 7. Correlation Heatmap

```python
corr_cols = ['age', 'num_prior_visits', 'length_of_stay', 'num_medications', 'num_comorbidities', 'readmitted_30days']
corr = df[corr_cols].corr()

fig, ax = plt.subplots(figsize=(6, 5))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='Blues', linewidths=0.5, ax=ax)
ax.set_title('Feature Correlation Heatmap', fontsize=11)
plt.tight_layout()
plt.savefig('../outputs/graphs/correlation_heatmap.png', dpi=150)
plt.show()
```

## 8. Prior Visits vs Readmission (Boxplot)

```python
fig, ax = plt.subplots(figsize=(5, 3.5))
data = [
    df[df['readmitted_30days'] == 0]['num_prior_visits'].values,
    df[df['readmitted_30days'] == 1]['num_prior_visits'].values,
]
bp = ax.boxplot(data, patch_artist=True, widths=0.4, medianprops=dict(color='white', linewidth=2))
bp['boxes'][0].set_facecolor('#4a7fa5')
bp['boxes'][1].set_facecolor('#b94a48')
ax.set_xticklabels(['Not Readmitted', 'Readmitted'])
ax.set_ylabel('Number of Prior Visits')
ax.set_title('Prior Visits vs Readmission', fontsize=11)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()
plt.savefig('../outputs/graphs/prior_visits_boxplot.png', dpi=150)
plt.show()
```
